```python
import torch
import transformers
```

```python
pipe = transformers.pipeline("text-generation", model="cognitivecomputations/dolphin-2_6-phi-2", trust_remote_code=True)
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant!"},
    {"role": "user", "content": """write a code in python to generate 1000 lines like: "i live in house 'number'" where the 'number' is a number of iteration. And put output in a file called output.txt"""},
]

prompt = pipe.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)
```

```python

terminators = [
    pipe.tokenizer.eos_token_id,
    pipe.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipe(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)

print(outputs[0]["generated_text"][len(prompt):])
```
